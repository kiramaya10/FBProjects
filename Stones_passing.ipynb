```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
df = pd.read_csv('/Users/kiramaya/Downloads/John_Stones_2022-2023.csv')
```

```python
df
```

```python
df['Pos'] = df['Pos'].str.split('\\',expand=True)[0]
```

```python
df.Pos.unique()
```

```python
df = df[(df['Pos']=='DM') | (df['Pos']=='CB')].reset_index()
```

```python
df
```

```python
df = df.drop(['index'],axis=1)
```

```python
df
```

```python
params = list(df.columns)
params = params[1:]
params
```

```python
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x])
    a = a - (a*.25)
    
    b = max(df[params][x])
    b = b + (b*.25)
    
    ranges.append((a,b))
    
for x in range(len(df['Pos'])):
    if df['Pos'][x] == 'DM':
        a_values = df.iloc[x].values.tolist()
    if df['Pos'][x] == 'CB':
        b_values = df.iloc[x].values.tolist()
        
a_values = a_values[1:]
b_values = b_values[1:]

values = [a_values,b_values]
```

```python
values
```

```python
title = dict(
    title_name='John Stones',
    title_color = 'red',
    subtitle_name = 'DM',
    subtitle_color = 'red',
    title_name_2='John Stones',
    title_color_2 = 'blue',
    subtitle_name_2 = 'CB',
    subtitle_color_2 = 'blue',
    title_fontsize = 18,
    subtitle_fontsize=15
)

endnote = '@kiramaya\ndata via FBREF'
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,
                         radar_color=['red','blue'],
                         alphas=[.75,.6],title=title,endnote=endnote,
                         compare=True)
```
